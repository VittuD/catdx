```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load predictions
predictions = pd.read_csv('predictions.csv')
```

```python
# Compute mean and std of the 'actual'
mean = predictions['actual'].mean()
std = predictions['actual'].std()
print('Mean:', mean)
print('Std:', std) 
```

```python
# Compute mean and std of the 'predicted'
mean = predictions['predicted'].mean()
std = predictions['predicted'].std()
print('Mean:', mean)
print('Std:', std)
```

```python
# Compute correlation to see if the prediction is just predicting the mean
correlation = predictions['actual'].corr(predictions['predicted'])
print('Correlation:', correlation)
```

```python
plt.scatter(predictions['actual'], predictions['predicted'], label='Predictions', color='blue')
plt.plot([predictions['actual'].min(), predictions['actual'].max()],
         [predictions['actual'].min(), predictions['actual'].max()],
         color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```
